```python
#goes through directory (yolov8 train, valid, or test), crops images by taking the center of the image and resizes the associated labels
new_height = 1024
new_width = 1024

directory = "../Downloads/20x2/train"
files = os.listdir(directory + "/images")

for image_name in files:
    #assumes folder in yolov8 format, where a folder has images and labels folder inside
    image_path = directory + "/images/" + image_name
    image = cv2.imread(image_path)
    old_height = image.shape[0]
    old_width = image.shape[1]
    left = int((old_width - new_width)/2)
    right =int((old_width + new_width)/2)
    top = int((old_height - new_height)/2)
    bottom = int((old_height + new_height)/2)
    cropped = image[top:bottom, left:right]
    cv2.imwrite(image_path, cropped)
    txt_path = directory + "/labels/" + image_name[:-3]+ "txt"
    file = open(txt_path, "r")
    print(file)
    boxes = file.readlines()
    print(boxes)
    for line in range(len(boxes)):
        print("Old:")
        print(boxes[line])
        numbers = boxes[line].split()
        #class_id center_x center_y width height - format of each line
        #calculate new numbers, write it into array
        old_center_x = float(numbers[1])
        old_center_y = float(numbers[2])
        old_w = float(numbers[3])
        old_h = float(numbers[4])
        new_center_x = float((old_center_x * old_width - left)/new_width)
        new_center_y = float((old_center_y * old_height - top)/new_height)
        if (new_center_x < 0 or new_center_y < 0 or new_center_x > 1 or new_center_y > 1):
            boxes[line] = ''
        else:
            height_scale = old_height/new_height
            width_scale = old_width/new_width
            new_h = old_h * height_scale
            new_w = old_w * width_scale
            if (new_center_x - new_w < 0):
                old_w = new_w
                old_centerx = new_center_x
                new_w = old_centerx + old_w/2
                new_center_x = new_w/2
            if (new_center_x + new_w > 1):
                old_w = new_w
                old_centerx = new_center_x
                new_w = 1-old_centerx + old_w/2
                new_center_x = (1 - new_w)/2
            if (new_center_y - new_h < 0):
                old_h = new_h
                old_centery = new_center_y
                new_h = old_centery + old_h/2
                new_center_y = new_h/2
            if (new_center_y + new_h > 1):
                old_h = new_h
                old_centery = new_center_y
                new_h = (1-old_centery) + old_h/2
                new_center_y = (1 - new_h)/2
            numbers[1] = str(new_center_x)
            numbers[2] = str(new_center_y)
            numbers[3] = str(new_w)
            numbers[4] = str(new_h)
            boxes[line] = numbers[0]+ " " + numbers[1] + " " + numbers[2] + " " + numbers[3] + " " + numbers[4]
            if len(boxes) > 1 and line < len(boxes) - 1:
                boxes[line] += '\n'
    file.close()
    file = open(txt_path, "w")
    file.writelines(boxes)
    file.close()
    #pull up the text file
    #write it
    
```

```python
#crops images and rewrites associated text files
#takes the first bounding box and chooses random boundaries that fully enclose the first bounding box
new_height = 640
new_width = 640

directory = "../Downloads/temp.v5i.yolov8/test"
files = os.listdir(directory + "/images")

for image_name in files:
    image_path = directory + "/images/" + image_name
    image = cv2.imread(image_path)
    old_height = image.shape[0]
    old_width = image.shape[1]
    print (old_height)
    print (old_width)
    #image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)\
    txt_path = directory + "/labels/" + image_name[:-3]+ "txt"
    file = open(txt_path, "r")
    boxes = file.readlines()
    #if no bounding boxes, just get the middle of the image
    if (len(boxes) == 0):
        left = int((old_width - new_width)/2)
        right = int((old_width + new_width)/2)
        top = int((old_height - new_height)/2)
        bottom = int((old_height + new_height)/2)
        cropped = image[top:bottom, left:right]
        cv2.imwrite(image_path, cropped)
    else:
        top_line = boxes[0]
        numbers = top_line.split()
        old_center_x = float(numbers[1])
        old_center_y = float(numbers[2])
        old_w = float(numbers[3])
        old_h = float(numbers[4])
        #get borders of bounding box in pixels on original image
        left_border = (old_center_x - old_w/2) * old_width
        right_border = (old_center_x + old_w/2) * old_width
        top_border = (old_center_y - old_h/2) * old_height
        bottom_border = (old_center_y + old_h/2) * old_height
        #get new image borders in pixels on original image
        new_left = random.randint(max(0,int(right_border - new_width)), min(int(left_border), old_width - new_width))
        new_right = new_left + new_width
        new_top = random.randint(max(0,int(bottom_border - new_height)),min(int(top_border), old_height - new_height))
        new_bottom = new_top + new_height
        print(f"Left: {new_left}")
        print(f"Right: {new_right}")
        print(f"top: {new_top}")
        print(f"Bottom: {new_bottom}")
        image = image[new_top:new_bottom,new_left:new_right]
        cv2.imwrite(image_path, image)
        new_h = old_h * old_height/new_height
        new_w = old_w * old_width/new_width
        new_x = float((old_center_x * old_width - new_left)/new_width)
        new_y = float((old_center_y * old_height - new_top)/new_height)
        numbers[1] = str(new_x)
        numbers[2] = str(new_y)
        numbers[3] = str(new_w)
        numbers[4] = str(new_h)
        boxes[0] = numbers[0]+ " " + numbers[1] + " " + numbers[2] + " " + numbers[3] + " " + numbers[4]
        if len(boxes) > 1:
            boxes[0] += '\n'
    line = 1
    while (line < len(boxes)):
        numbers = boxes[line].split()
        #class_id center_x center_y width height - format of each line
        #calculate new numbers, write it into array
        old_center_x = float(numbers[1])
        old_center_y = float(numbers[2])
        old_w = float(numbers[3])
        old_h = float(numbers[4])
        # boundaries in pixels of box on old image
        left = (old_center_x - old_w/2) * old_width
        right = (old_center_x + old_w/2) * old_width
        top = (old_center_y - old_h/2) * old_height
        bottom = (old_center_y + old_h/2) * old_height
        new_center_x = float((old_center_x * old_width - new_left)/new_width)
        new_center_y = float((old_center_y * old_height - new_top)/new_height)
        if (new_center_x < 0 or new_center_y < 0 or new_center_x > 1 or new_center_y > 1):
            boxes[line] = ''
        else:
            height_scale = old_height/new_height
            width_scale = old_width/new_width
            new_h = old_h * height_scale
            new_w = old_w * width_scale
            if (new_center_x - new_w < 0):
                old_w = new_w
                old_centerx = new_center_x
                new_w = old_centerx + old_w/2
                new_center_x = new_w/2
            if (new_center_x + new_w > 1):
                old_w = new_w
                old_centerx = new_center_x
                new_w = 1-old_centerx + old_w/2
                new_center_x = (1 - new_w/2)
            if (new_center_y - new_h < 0):
                old_h = new_h
                old_centery = new_center_y
                new_h = old_centery + old_h/2
                new_center_y = new_h/2
            if (new_center_y + new_h > 1):
                old_h = new_h
                old_centery = new_center_y
                new_h = (1-old_centery) + old_h/2
                new_center_y = (1 - new_h/2)
            numbers[1] = str(new_center_x)
            numbers[2] = str(new_center_y)
            numbers[3] = str(new_w)
            numbers[4] = str(new_h)
            boxes[line] = numbers[0]+ " " + numbers[1] + " " + numbers[2] + " " + numbers[3] + " " + numbers[4]
            if len(boxes) > 1 and line < len(boxes) - 1:
                boxes[line] += '\n'
        line +=1
    file.close()
    file = open(txt_path, "w")
    file.writelines(boxes)
    file.close()
    #pull up the text file
    #write it
```
